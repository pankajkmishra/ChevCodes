# Plot the Marmousi velocity model
We will use this model to demonstrate different ways of performaning seismic modeling.

```julia
using PyPlot # Add matplotlib for visualizations 
using MAT    # Do MATLAB manipulations
```

```julia
file = matopen("vp.mat")
v = read(file, "v")
close(file)
```

```julia
figure(figsize=(10,5)); clf()
imshow(v, aspect="1",cmap = "seismic", extent =[0,35,15,0]);
colorbar();
title("SEAM_vp (m/s)");
```
